```python
# to find the odd and even numbers from a list

list1=[50,60,99,41,23,87,5,44,99,25,1,70]
even_no=[]
odd_no=[]

for i in list1:
  if i%2==0:
    #print(i, "is even number")
    even_no.append(i)
  else:
    #print(i, "is odd number")
    odd_no.append(i)

print("List of even_no is:", even_no,"count of even numbers",  len(even_no))
print("List of odd_no is:", odd_no,"count of odd numbers",  len(odd_no))

```

```python
# create a dictionary of 5 students and get the total marks of each students in a dictionary.
# find the topper studens and their marks
student_data={
    'stu1':{
    'name':'preeti',
    'age':24,
    'subjects':{
        'english':80,
        'maths':90,
        'science':85,
        'computers':95
    },
    'year':2022
},

'stu2':{
    'name':'priya',
    'age':22,
    'subjects':{
        'english':70,
        'maths':60,
        'science':55,
        'computers':75
    },
    'year':2022
}
}
print(student_data)

```

```python
student_marks={}

for i, j in student_data.items():
  for x,y in j.items():
    if x=='subjects':
      student_marks[i]=sum(y.values())
      
print(student_marks)

topper=max(student_marks, key=student_marks.get)
print("Topper of the class is:", topper, "and their marks is: ", student_marks[topper])
```

```python
student_marks={}

for i, j in student_data.items():
  for x,y in j.items():
    if x=='subjects':
      #student_marks[i]=sum(y.values())
      student_marks[i]=sum(y.values())/400*100

      
print(student_marks)

topper=max(student_marks, key=student_marks.get)
print("Topper of the class is:", topper, "and their percentage is: ", student_marks[topper])
```

```python
# positional arguements
def my_func(name,age,city):
  print('name =', name)
  print('age =', age)
  print('city =', city)

my_func('preeti', 'Delhi',30)

print('---------------------------')

# keyword arguments
def my_func(name,age,city):
  print('name =', name)
  print('age =', age)
  print('city =', city)
  
my_func(name='preeti', city='Delhi', age=30)

```

```python
#variable-length positional arguments
def my_func(*args):
  for i in args:
    print(i)
  
my_func('preeti', 'Delhi',30, 2022, 'python')

print('------------------------------')

# variable-length keyword arguments
def my_func1(**kwargs):
  for key,value in kwargs.items():
    print(key,value)

my_func1(name='preeti', city='Delhi', age=30, year=2022, course='python')
```

```python
class person:
  def __init__(self,name,age,contact,email):
    self.name=name
    self.age=age
    self.contact=contact
    self.email=email

  def greet(self):
    print("Hello ", self.name, "Welcome to my world")
    print(self.email, self.contact)

p=person("pushpa", 25, 9876543210, 'abc@gmail.com')

print(p.name)
print(p.age)
#p.greet()
```

```python

class student(person):
  pass

s=student("komal",'sharma','xyz@gmail.com',9810262545)
s.greet()
```
